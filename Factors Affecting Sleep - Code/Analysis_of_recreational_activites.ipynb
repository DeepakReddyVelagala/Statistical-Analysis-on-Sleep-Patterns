```python
def randomint(num):
  a=num*1000
  if a==0:
    return 0
  else:
    b=(num-1)*1000
    num3 = random.randint(b,a)
    return float(num3/1000)
```

## Normal distribution graph and confidence intervals

### Graph

```python
x_axis = np.arange(3, 11, 0.01)
plt.plot(x_axis, norm.pdf(x_axis, df["How much sleep do you get in a day?"].mean(), df["How much sleep do you get in a day?"].std()),label='for all kind of activities and sports')
for i in dff:
  plt.plot(x_axis, norm.pdf(x_axis, i["How much sleep do you get in a day?"].mean(), i["How much sleep do you get in a day?"].std()))
plt.legend()
plt.show()
```

```python
means_40,means_50,means_60,means_70,means_80,means_90=[],[],[],[],[],[]
for i in range(sample_size):
  means_40.append(sample_of_size_40[i].mean())
  means_50.append(sample_of_size_50[i].mean())
  means_60.append(sample_of_size_60[i].mean())
  means_70.append(sample_of_size_70[i].mean())
  means_80.append(sample_of_size_80[i].mean())
  means_90.append(sample_of_size_80[i].mean())
```
